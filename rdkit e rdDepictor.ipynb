```python
# Importar as bibliotecas necess√°rias
from rdkit import Chem
from rdkit.Chem import rdDepictor
#from rdkit.Chem.Draw import IPythonConsole
```

```python
rdDepictor.SetPreferCoordGen(True)
benzeno_string = 'c1ccccc1'
benzeno_molecula = Chem.MolFromSmiles(benzeno_string)
rdDepictor.Compute2DCoords(benzeno_molecula)
benzeno_molecula
```

```python
rdDepictor.SetPreferCoordGen(True)
beta_caroteno_string = 'CC1CCC/C(C)=C1/C=C/C(C)=C/C=C/C(C)=C/C=C/C=C(C)/C=C/C=C(C)/C=C/C2=C(C)/CCCC2(C)C'
beta_caroteno_molecula = Chem.MolFromSmiles(beta_caroteno_string)
rdDepictor.Compute2DCoords(beta_caroteno_molecula)
beta_caroteno_molecula
```

```python
rdDepictor.SetPreferCoordGen(True)
sulfato_cobre_string = '[Cu+2].[O-]S(=O)(=O)[O-]'
sulfato_cobre_molecula = Chem.MolFromSmiles(sulfato_cobre_string)
rdDepictor.Compute2DCoords(sulfato_cobre_molecula)
sulfato_cobre_molecula
```
